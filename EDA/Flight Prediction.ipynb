### Handling date values

```python
df["Date"] = df['Date_of_Journey'].str.split("/").str[0]
df["Month"] = df['Date_of_Journey'].str.split("/").str[1]
df["Year"] = df['Date_of_Journey'].str.split("/").str[2]
df.head()
```

```python
df.Date = df.Date.astype(int)
df.Month = df.Month.astype(int)
df.Year = df.Year.astype(int)
df.info()
```

```python
#Drop Date of Journey Column
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
# Handling Arrival Time Column
df.Arrival_Time = df.Arrival_Time.apply(lambda x:x.split(" ")[0])
df.head()
```

```python
df["Arrival_hour"] = df.Arrival_Time.apply(lambda x:x.split(":")[0])
df["Arrival_min"] = df.Arrival_Time.apply(lambda x:x.split(":")[1])
df.head()
```

```python
df.Arrival_hour = df.Arrival_hour.astype(int)
df.Arrival_min = df.Arrival_min.astype(int)
df.drop("Arrival_Time", axis=1, inplace=True)
```

```python
df.head()
```

### Handling Departure Time Column

```python
df["Deptarture_Hour"] = df.Dep_Time.apply(lambda x:x.split(":")[0])
df["Deptarture_Min"] = df.Dep_Time.apply(lambda x:x.split(":")[1])
df.head()
```

```python
df.Deptarture_Hour = df.Deptarture_Hour.astype(int)
df.Deptarture_Min = df.Deptarture_Min.astype(int)
df.drop("Dep_Time", axis=1, inplace=True)
```

```python
df.head()
```

### Handling Duration Column


```python
def min_value(x):
    val = x.split(" ")
    if len(val)>1:
        return val[1].split("m")[0]
    return 0
```

```python
df['Duration_Hour'] = df.Duration.apply(lambda x:x.split(" ")[0]).apply(lambda x:x.split("h")[0])
#df["Duartion_Minute"] = df.Duration.apply(lambda x:x.split(" ")[1] if len(x.split(" ")>0 else 0)).apply(lambda x:x.split("m")[0])
df.head()
```

```python
df.drop("Duartion_Minute", axis=1, inplace=True)
df.head()
```

```python
df["Duartion_Minute"] = df.Duration.apply(min_value)
df.head()
```
