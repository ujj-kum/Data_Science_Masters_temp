# Feature Scaling

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
sns.histplot(df.total_bill, kde=True)
```

```python
# Apply Standardization on Total_BIll
```

```python
mean = df.total_bill.mean()
std = df.total_bill.std()
mean, std
```

```python
df['total_bill_standardized'] = (df.total_bill-mean)/std
```

```python
sns.histplot(df.total_bill_standardized, kde=True)
```

```python
#Using sklearn
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df[['total_bill']])
scaler.transform(df[['total_bill']])
```

# Data Encoding

### 1.One Hot Encoding/Nominal Encoding

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame(
{
    "Color":["Red", "Blue", "Green", "Green",
            "Red", "Blue"]
}
)
```

```python
df.head()
```

```python
df.Color.value_counts()
```

```python
encoder = OneHotEncoder(drop='first')
```

```python
encoder.fit(df[["Color"]])
```

```python
encoded = encoder.transform(df[["Color"]]).toarray()
```

```python
df_encoded = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```

```python
df_encoded
```

```python
#For new data
encoder.transform([['Blue']]).toarray()
```

```python
pd.concat([df, df_encoded], axis=1)
```

### Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
encoder.fit_transform(df[['Color']])##
```

### Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({"Size":["Small", "Medium", "Large", "Medium", "Small", "Large"]})
```

```python
df
```

```python
encoder = OrdinalEncoder(categories=[df.Size.unique()])
```

```python
encoder.fit_transform(df[['Size']])
```

### Target Guided Ordinal Encoding

```python
df = pd.DataFrame({
    "City":["New York", "London", "Paris", "Tokyo", "New York", "Paris"],
    "Price":[200, 150, 300, 250, 180, 320]
})
df
```

```python
mean_price = df.groupby('City')['Price'].mean().to_dict()
mean_price
```

```python
df['City_Encoded'] = df.City.map(mean_price)
df
```
