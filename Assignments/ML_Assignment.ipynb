# 1. Logistic Regression

```python
x = [random.random()*10 for _ in range(10)]
y = [random.random()*10 for _ in range(10)]
classes = [0, 1]
category = [random.choice(classes) for _ in range(10)]
df_dict = {"X":x, "Y":y, "Actual_Output":category}
```

```python
df = pd.DataFrame(df_dict)
df
```

### Formula for Logistic Regression

$$ z = w_0 \cdot x + w_1 \cdot y + b $$

The sigmoid function is given by: $$ sigma(z) = \frac{1}{1 + e^{-z}} $$


```python
def sigmoid(z):
  return 1/1+np.exp(-z)

def output_update_param(y, x1, w1, x2, w2, b):
    alpha = 0.001
    pred  = sigmoid(x1*w1+x2*w2+b)
    w1 = w1+alpha*(y-pred)*pred*(1-pred)*x1
    w2 = w2+alpha*(y-pred)*pred*(1-pred)*x2
    b = b+alpha*(y-pred)*pred*(1-pred)*1
    return [w1, w2, b]
```

```python
w1, w2, b = 0, 0, 0
weights_final = []
for i in range(10):
    y = df.iloc[i, 2]
    x1 = df.iloc[i, 0]
    x2 = df.iloc[i, 1]
    weights = output_update_param(y, x1, w1, x2, w2, b)
    weights_final.append(weights)
```

```python
weights_final_df = pd.DataFrame(weights_final, columns=['w1', 'w2', 'b'])
weights_final_df
```

```python
df = pd.concat([df, weights_final_df], axis=1)
df
```
