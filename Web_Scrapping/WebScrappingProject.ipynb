```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
from urllib import request
import logging
```

```python
main_url = 'https://www.amazon.in/s?k='
item_name = "tv"
amazon_url = main_url+item_name
amazon_url
```

```python
import urllib.request
from urllib.error import HTTPError

url = amazon_url

try:
    # Open the URL and read its content
    with urllib.request.urlopen(url) as response:
        html = response.read()
        print(html)  # This will print the HTML content of the website
except HTTPError as e:
    print(f"HTTP Error: {e.code} {e.reason}")
```

### Using Beautiful Soup

```python
amazon_html = bs(html, 'html.parser')
```

```python
# For big box data

bigbox = amazon_html.find_all("div", {'class':'a-section'})
len(bigbox)
```
