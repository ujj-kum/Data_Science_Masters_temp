### Plotting 3D Graphs

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
plt.scatter(x,y,z)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x,y,z,c='r',alpha=1)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```
