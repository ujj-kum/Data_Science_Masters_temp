### Merge

```python
d1 = pd.DataFrame(d)
d1.head()
```

```python
d2 = pd.DataFrame({'Key1': [27, 3, 48, 5], 'Key2': [46, 57, 6, 7], 'Key3': [8, 98, 109, 11]})
d2.head()
```

```python
pd.merge(d1,d2, how='left')
```

```python
pd.merge(d1,d2, how='right')
```

```python
pd.merge(d1,d2, how='outer')
```

```python
pd.merge(d1,d2, how='cross')
```

```python
#Join checks for indexes
d1.join(d2, lsuffix="_left", rsuffix="_right")
```

```python
#Join checks for indexes
d1.join(d2, how="inner",lsuffix="_left", rsuffix="_right")
```

### Apply

```python
#Creating new Columns based on seggeregation
df["Fare_INR"] = df.Fare.apply(lambda x: x*80)
df.head()
```

```python
df["len_name"] = df.Name.apply(lambda x:len(x))
df.head()
```

```python
def flag(x):
    if x<10:
        return "Cheap"
    elif x>10 and x<20:
        return "Mid"
    else:
        return "High"
```

```python
df['Flag_fare'] = df['Fare'].apply(flag)
df.head()
```

```python
df.groupby("Flag_fare").mean()
```

### Date Functions

```python
df_date.dtypes
```

```python
df7 = pd.DataFrame({"Date":df_date.iloc[0:6,0]})
df7.head()
```
