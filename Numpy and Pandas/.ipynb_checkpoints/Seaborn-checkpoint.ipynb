```python
tips = sns.load_dataset('tips')
tips
```

X axis = Total Bill
Y axis = Total Tip


```python
sns.scatterplot(x=tips.total_bill, y=tips.tip)
```

Smoker V/s Non Smoker

```python
sns.displot(tips.smoker)
```

```python
tips['smoker'].value_counts()
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip, data=tips, hue='smoker')
```

```python
tips['size'].value_counts()
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,hue='size')
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,style='size')
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,style='size',hue='time')
```

Day VS Total Bill

```python
sns.catplot(x='day', y='total_bill', data=tips)
```

```python
sns.displot(tips.day)
```

```python
sns.jointplot(x=tips.total_bill,y=tips.tip)
```

```python
sns.pairplot(tips)
```
